```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./tap_fun_train.csv')
```

```python
df2 = df[['wood_add_value', 'wood_reduce_value',
       'stone_add_value', 'stone_reduce_value', 'ivory_add_value',
       'ivory_reduce_value', 'meat_add_value', 'meat_reduce_value','pve_battle_count', 'pve_lanch_count', 'pve_win_count',
       'avg_online_minutes', 'pay_price', 'pay_count']]
```

```python
df.columns
```

```python
df.prediction_pay_price[df['prediction_pay_price']!=0]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
```

```python
model.fit(df2,df.prediction_pay_price)
```

```python
train_score = model.score(df2,df.prediction_pay_price)
```

```python
train_score
```

```python
test = pd.read_csv('./tap_fun_test.csv')
```

```python
test
```

```python
test_pre = model.predict(test[['wood_add_value', 'wood_reduce_value',
       'stone_add_value', 'stone_reduce_value', 'ivory_add_value',
       'ivory_reduce_value', 'meat_add_value', 'meat_reduce_value','pve_battle_count', 'pve_lanch_count', 'pve_win_count',
       'avg_online_minutes', 'pay_price', 'pay_count']])
```

```python
test_pre[test_pre<0]=0
```

```python
test_pre
```

```python
dfsa1 = pd.DataFrame()
dfsa1['user_id']= test.user_id
dfsa1['prediction_pay_price'] = test_pre
```
